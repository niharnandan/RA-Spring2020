### Create plots and save them as pdf

```python
import matplotlib.pyplot as plt
import matplotlib.style as style
import plot
import seaborn as sns
import json
style.use('seaborn')
plt.rcParams.update({'font.size': 10})
```

```python
# Open JSON file to load data
with open('data.json') as f:
    plot_json = json.load(f)
```

### Loop to go over elements in the JSON file and plot based on plot type.

```python
for i in plot_json:
    if i == 'bar':
        data = plot_json[i]
        plot.bar(data['title'], data['t2_perc_confirmatory_round1'], data['xlabel'], data['ylabel'], data['colors'], 
           data['names'], data['yaxis'], data['xtick'], data['ytick'], data['width'])
    elif i == 'scatter':
        data = plot_json[i]
        plot.scatter(data['title'], data["t2_perc_confirmatory_allrounds"], data["t2_avg_samples"], \
                 data['xlabel'], data['ylabel'], data["xaxis"], data["yaxis"], data["colors"], \
                 data["xtick"], data["ytick"])
```
