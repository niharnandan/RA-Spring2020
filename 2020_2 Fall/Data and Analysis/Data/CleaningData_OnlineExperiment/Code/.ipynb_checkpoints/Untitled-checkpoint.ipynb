```python
def plot2d(title, x, y, xlabel, ylabel, colors=['red'], lw=[1], ls=['-.'], xaxis='', yaxis='', xtick='', ytick=''):
    for i in range(len(y)):
        plt.plot(x, y[i], linewidth=lw[i], color = colors[i], linestyle=ls[i])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
```

```python
plot2d('Title', list(range(1,10)), [list(range(1,10))], 'x', 'y')
```

```python
def bar(title, y, xlabel, ylabel, colors=['red','blue'], names = [''], yaxis='', xtick='', ytick=''):
    x = [i*1.5 for i in range(len(y))]
    plt.bar(x,y, color=colors, edgecolor='black', tick_label=names) 
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
```

```python
bar('Title', list(range(1,3)), 'x', 'y', ['red','blue'], ['human','bayes'])
```
